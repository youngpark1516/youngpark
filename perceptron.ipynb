```python
def OR(x1,x2):
    while True:
        ow1 = np.random.normal()
        ow2 = np.random.normal()
        ob = np.random.normal()
        if(1*ow1+1*ow2+ob <= 0):
            pass
        elif(0*ow1+1*ow2+ob <= 0):
            pass
        elif(1*ow1+0*ow2+ob <= 0):
            pass
        elif(0*ow1+0*ow2+ob > 0):
            pass
        else:
            break
    w = ow1
    x = ow2
    c = ob
    if(x1*w+x2*x+c <= 0):
        return 0
    else:
        return 1
def NAND(x1,x2):
    while True:
        naw1 = np.random.normal()
        naw2 = np.random.normal()
        nab = np.random.normal()
        if(1*naw1+1*naw2+nab > 0):
            pass
        elif(0*naw1+1*naw2+nab <= 0):
            pass
        elif(1*naw1+0*naw2+nab <= 0):
            pass
        elif(0*naw1+0*naw2+nab <= 0):
            pass
        else:
            break
    w = naw1
    x = naw2
    c = nab
    if(x1*w+x2*x+c <= 0):
        return 0
    else:
        return 1
def AND(x1,x2):
    while True:
        aw1 = np.random.normal()
        aw2 = np.random.normal()
        ab = np.random.normal()
        if(1*aw1+1*aw2+ab <= 0):
            pass
        elif(0*aw1+1*aw2+ab > 0):
            pass
        elif(1*aw1+0*aw2+ab > 0):
            pass
        elif(0*aw1+0*aw2+ab > 0):
            pass
        else:
            break
    w = aw1
    x = aw2
    c = ab
    if(x1*w+x2*x+c <= 0):
        return 0
    else:
        return 1
def XOR(x1,x2):
    return AND(NAND(x1,x2),OR(x1,x2))
```
