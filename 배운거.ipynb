```python
import requests
import bs4

h = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}
p_url = 'https://www.pubg.com/category/patch-notes-en-gb/'
p_request_result = requests.get(p_url, headers = h)
pubg = p_request_result.text

p_soup = bs4.BeautifulSoup(pubg, 'html')
print(p_soup)
```

```python
def get_source(url):
    h = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}
    request_result = requests.get(url, headers = h)
    game = request_result.text
    soup = bs4.BeautifulSoup(game, 'html')
    return soup
```
