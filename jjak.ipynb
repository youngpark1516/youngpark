```python
for i in range(1,100):
    jjak = False
    jjakx2 = False
    if i % 3 == 0:
        jjak = True
    if str(i)[len(str(i))-1] == '3' or str(i)[len(str(i))-1] == '6' or str(i)[len(str(i))-1] == '9':
        jjakx2 = True
    if jjak and jjakx2:
        print(" ")
    elif jjak:
        print("jjak")
    elif jjakx2:
        print("jjakjjak")
    else:
        print(i)
```
