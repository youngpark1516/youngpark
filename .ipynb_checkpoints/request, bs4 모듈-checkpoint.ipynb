## 영화 랭킹 불러오기

```python
#준비
import requests #모듈 불러오기
url2 = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn' #사이트 소스 가져오기
request_result = requests.get(url2) 
naver_movie = request_result.text #소스를 문자열로 바꾸기

#코드
l = [] #임시 리스트
sajun = {} #사전
word = '<td class="title">\r\n\t\t\t\t\t\t<div class="tit3">\r\n\t\t\t\t\t\t\t<a href="/movie/bi/mi/basic.nhn?code='
word2 = '</a>\r\n\t\t\t\t\t\t</div>\r\n\t\t\t\t\t</td>\r\n\t\t\t\t\t<!-- 평점순일 때 평점 추가하기  -->'
for i in range(1,naver_movie.count(word)+1): #등수만큼 반복
    l.append(naver_movie.split(word)[i].split(word2)[0].split('>')[1]) #리스트에 차례로 넣기
    sajun[i]=l[i-1] #사전에 입력
print(sajun)
```

```python
def bullaogi(sutja=1): #기본 매개 변수
    number = int(sutja) #수치형으로 변경
    print(sajun[number]) #입력된 숫자에 맞는 등수의 영화 프린트
```

```python
bullaogi(40) #원하는 등수 입력 (만약 입력x, 1등이 나옴)
```

```python
import requests
import bs4
url2 = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
request_result = requests.get(url2)
naver_movie = request_result.text

soup = bs4.BeautifulSoup(naver_movie, 'html')
```

```python
print(soup)
```

```python
soup.find('div', attrs = {'class':'tit3'}).text.strip()
```

```python
for i in soup.find_all('div', attrs = {'class':'tit3'}):
    print(i.text.strip())
```

```python
import requests 
import bs4
url = 'https://www.worldometers.info/coronavirus/' #사이트 소스 가져오기
request_result = requests.get(url) 
corona = request_result.text #소스를 문자열로 바꾸기

soup = bs4.BeautifulSoup(corona,'html')
temp = soup.find('tr',class_='total_row_world')
for i in range(1,100):
    temp = temp.findNext('tr')
    if(temp.find('a',class_='mt_a') != None):
        print(i+1 , '국가 : ', temp.find('a',class_='mt_a').text)
        print('총 확진자 수 :', temp.findNext('td').findNext('td').text)
        if (temp.findNext('td').findNext('td').findNext('td').text == ''):
            print('새로운 확진자 수: ',0)
        else:
            print('새로운 확진자 수: ',temp.findNext('td').findNext('td').findNext('td').text)
        print()
```
