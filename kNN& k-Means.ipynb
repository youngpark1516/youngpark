```python
import random
import math
def kNN(h, w, r, k):
    d = []
    for i in range(200):
        d.append([math.sqrt(pow(p[i][0]-int(h),2)+pow(p[i][1]-int(w),2)+pow(p[i][2]-int(r),2)),p[i][3]])
    d.sort()
    print(d[0:k])
    countf = 0
    countm = 0
    for i in range(k):
        if(d[i][1]==0):
            countf += 1
        else:
            countm += 1
    if(countf>countm):
        print('you are a female')
    elif(countm>countf):
        print('you are a male')
```

### k-Means

```python
p = []
for i in range(100):
    p.append([random.randrange(160,200,1),random.randrange(60,100,1)])
    p.append([random.randrange(140,180,1),random.randrange(40,80,1)])
print(p)
```

```python
center1 = [140,70]
center2 = [150,70]
```

```python
import matplotlib.pyplot as plt
import random
%matplotlib inline

ts = []

for i in range(10):
    temp1 = []
    temp2 = []
    for i in range(200):
        if (math.sqrt(pow(p[i][0]-center1[0],2)+pow(p[i][1]-center1[1],2)) > math.sqrt(pow(p[i][0]-center2[0],2)+pow(p[i][1]-center2[1],2))):
            temp1.append(p[i])
        else:
            temp2.append(p[i])
    temp1sum_h = 0
    temp1sum_w = 0
    temp2sum_h = 0
    temp2sum_w = 0
    for i in range(len(temp1)):
        temp1sum_h += temp1[i][0]
        temp1sum_w += temp1[i][1]

    for i in range(len(temp2)):
        temp2sum_h += temp2[i][0]
        temp2sum_w += temp2[i][1]
    newx1 = temp1sum_h/len(temp1)
    newy1 = temp1sum_w/len(temp1)
    newx2 = temp2sum_h/len(temp2)
    newy2 = temp2sum_w/len(temp2)
    center2 = [newx1, newy1]
    center1 = [newx2, newy2]
    print(center1, center2)
    for i in temp1:
        plt.plot(i[0],i[1],'o',color='r')
    for i in temp2:
        plt.plot(i[0],i[1],'o',color='b')
        
    ts.append([temp1,temp2])
        
temp1.sort()
temp2.sort()
print(temp1)
print(temp2)
```
