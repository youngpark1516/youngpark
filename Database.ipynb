```python
import requests 
import bs4
import time
import schedule
import pymysql
db = pymysql.connect(host='127.0.0.1', #내 주소
                     port=3306, user='young', 
                     passwd = '1234', 
                     db='corona', charset='utf8')
cursor = db.cursor()
url = 'https://www.worldometers.info/coronavirus/' #사이트 소스 가져오기

def job():

    request_result = requests.get(url) 
    corona = request_result.text #소스를 문자열로 바꾸기
    soup = bs4.BeautifulSoup(corona,'html')
    temp = soup.find('tr',class_='total_row_world')
    for i in range(1,108):
        temp = temp.findNext('tr')
        if(temp.find('a',class_='mt_a') != None):
            nara =  temp.find('a',class_='mt_a').text
            total_infected = temp.findNext('td').findNext('td').text.replace(',','')
            if (temp.findNext('td').findNext('td').findNext('td').text == ''):
                new_infected = 0
            else:
                new_infected = temp.findNext('td').findNext('td').findNext('td').text.replace(',','').replace('+','')
            if(temp.findNext('td').findNext('td').findNext('td').findNext('td').text == ' '):
                total_dead = 0
            else:
                total_dead = temp.findNext('td').findNext('td').findNext('td').findNext('td').text.replace(",",'')
            if (temp.findNext('td').findNext('td').findNext('td').findNext('td').findNext('td').text == ''):
                new_dead = 0
            else:
                new_dead = temp.findNext('td').findNext('td').findNext('td').findNext('td').findNext('td').text.replace(',','').replace('+','')
            sql = "Insert into corona.corona_case(country, total_case, new_case, total_death, new_death) Values("+"'"+nara+"'"+","+str(total_infected)+","+str(new_infected)+","+str(total_dead)+","+str(new_dead)+");"
            cursor.execute(sql)
            db.commit()
    
schedule.every().day.at("07:00").do(job)

while True:
    schedule.run_pending()
    time.sleep(1)
```
